```python
import pandas as pd
score = pd.read_csv('data/score.csv', index_col='지원번호')
score
```

```python
df_score = score.loc[:, '국어':'사회']
df_score
```

```python
df=df_score.copy()
df.reset_index(inplace=True)
df['지원번호'] = df['지원번호'].str.replace('번','')
df.set_index('지원번호',inplace=True)
df.to_csv('data/학생성적.csv')
```

```python
info= pd.read_csv('data/학생정보.csv', index_col='지원번호')
info
```

```python
score = pd.read_csv('data/학생성적.csv', index_col='지원번호')
score
```

```python
merge = pd.merge(score, info, how='right', on='지원번호')
merge.set_index('지원번호', inplace=True)
merge.index
```

```python
merge.columns
```

```python
df=info.join(score)
df
```

```python

df['총점']=df.apply(lambda row:sum(row['국어':'사회']), axis=1)
df['평균']=df['총점']/5
df
```

```python
no=2
row = df.loc[no]

print(f'지원번호:{no} 이름:{row["이름"]} 학교:{row["학교"]} 평균:{row["평균"]:.2f}')
```

```python
word = '태'
filt = df['이름'].str.contains(word)
idxs = df[filt].index
for idx in idxs:
    row = df.loc[idx]
    print(f'지원번호:{idx} 이름:{row["이름"]} 학교:{row["학교"]} 평균:{row["평균"]:.2f}')
```

```python
word = '북산'
filt = df['학교'].str.contains(word)
idxs = df[filt].index
for idx in idxs:
    row = df.loc[idx]
    print(f'지원번호:{idx} 이름:{row["이름"]} 학교:{row["학교"]} 평균:{row["평균"]:.2f}')
```
